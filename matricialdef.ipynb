```octave
clear all

A = [1:3; 4:6; 7:9; 10:12; 13:15]

B = NonsingularMat(3)
C = NonsingularMat(5)

CA = C*A
AB = A*B

display("A:\n")
rA = rank(A)
printf("Rango: %s\n", num2str(rA))

display("CA:\n")
rCA = rank(CA)
printf("Rango: %s\n", num2str(rCA))

display("\nAB:")
rAB = rank(AB);
printf("Rango: %s", num2str(rAB))
```

Los rangos no cambian pues al ser B y C no singulares su producto por otra matriz no cambia el rango de esta última.
